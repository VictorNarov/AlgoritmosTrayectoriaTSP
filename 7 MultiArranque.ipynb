# Métodos MultiArranque
## GRASP

```python
import import_ipynb
from algoritmos import *

semilla = random.randint(0,9999999)
```

#### GRASP con selección probabilística (inversa a la distancia)

```python
fichero = "st70"
print(fichero)
#,"a280","p654","vm1084","vm1748"
nodos = leeFichero(fichero)
distancias = getMDistancias(nodos)
numNodos = len(distancias[1])
tamaLRC = math.floor(0.1 * numNodos)

t = time.time()

[mejorCamino, mejorCoste, ev]  = GRASP(distancias, tamaLRC, semilla, True)

elapsed = time.time() - t


print("---\nMejor Solución Coste:\t%i" % mejorCoste)
print("EV:\t\t\t%i" % ev)
print("Semilla:\t\t%i" % semilla)
print("Tiempo de ejecución:\t%f\n" % elapsed )


titulo = "GRASP" + " - Mejor solución " + fichero + ": " + str(mejorCoste)
pintaCamino(mejorCamino, nodos, titulo)
```
