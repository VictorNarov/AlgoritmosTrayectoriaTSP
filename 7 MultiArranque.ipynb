#### GRASP con selección probabilística (inversa a la distancia)

```python
fichero = "st70"
#,"a280","p654","vm1084","vm1748"
nodos = leeFichero(fichero)
distancias = getMDistancias(nodos)
numNodos = len(distancias[1])
tamaLRC = math.floor(0.1 * numNodos)

[mejorCamino, mejorCoste]  = GRASP(distancias, tamaLRC, random.randint(0,9999999))

print("---\nMejor Solución Coste: %i" % mejorCoste)
titulo = "GRASP" + " - Mejor solución " + fichero + ": " + str(mejorCoste)
pintaCamino(mejorCamino, nodos, titulo)
```
