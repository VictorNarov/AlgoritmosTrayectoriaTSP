## Fichero st70

```python
algoritmo = "algoritmoGreedy"
print(algoritmo)
fichero = "st70"
#,"a280","p654","vm1084","vm1748"
nodos = leeFichero(fichero)
distancias = getMDistancias(nodos)
   

t = time.time()
with Pool(processes=10) as p:
     camino = p.starmap(algoritmoGreedy, product([distancias]))[0]


elapsed = time.time() - t
print("Tiempo de ejecuci√≥n: %f\n" % elapsed )

coste = getCosteCamino(camino,distancias)

titulo = algoritmo + " - Mejor soluci√≥n " + fichero + ": " + str(coste)
print("Coste: %s" % (coste) ) 
#print("Camino: " + str(camino))
pintaCamino(camino,nodos,titulo)
```
