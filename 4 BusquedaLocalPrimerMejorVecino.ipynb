# Búsqueda Local Primer Mejor Vecino

```python
from ipynb.fs.full.algoritmos import *
```

## Generación de semillas aleatorias

```python
semillas = []
for i in range(10):
    semillas.append(random.randint(0,9999999))
```

## Fichero st70

```python
algoritmo = "busquedaLocalPrimerMejorVecino"
print(algoritmo)
fichero = "st70"
#,"a280","p654","vm1084","vm1748"
nodos = leeFichero(fichero)
distancias = getMDistancias(nodos)
nIteraciones = 1600 * len(nodos)
    
print(f"\nFichero: {fichero}\nLimite iteraciones: {nIteraciones}")

costes = []
evs = []
mejorSolucion = []
mejorCoste = float('inf')

t = time.time()
with Pool(processes=10) as p:
     resultados = p.starmap(busquedaLocalPrimerMejorVecino_v2, product([distancias], semillas, [nIteraciones]))



#for s in semillas:
    
#resultados = [busquedaLocalMejorVecino_v2(distancias, 1,nIteraciones)]
    
elapsed = time.time() - t
print("Tiempo de ejecución: %f\n" % elapsed )

i=0
for res in resultados:
    costes.append(res[1])
    evs.append(res[2])
    
    if res[1] < mejorCoste:
        mejorCoste = res[1] 
        mejorSolucion = res[0]

    #coste = getCosteCamino(camino, distancias)
    print("%i)\tCoste: %s \t #EV: %i \t Semilla: %s" % (i+1,res[1], res[2], res[3]) ) 
    #print("Camino: " + str(camino))
    # pintaCamino(camino,nodos)
    i+=1

print("-----\nCoste Media:\t%f \t Desv: %f" % (mean(costes), stdev(costes)))
print("#Ev Media:\t%f \t Desv: %f" % (mean(evs), stdev(evs)))
print("Mejor solución de coste: ", str(mejorCoste))
titulo = algoritmo + " - Mejor solución " + fichero + ": " + str(mejorCoste)
pintaCamino(mejorSolucion, nodos, titulo)
```
